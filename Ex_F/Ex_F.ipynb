# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import torch
#from fairseq.models.data2vec import Data2VecTextModel
```

# Load Data Set

```python
file_path = "../DB/DMN Report 29.xlsx"

xls = pd.ExcelFile(file_path)
column_array=pd.read_excel(file_path, sheet_name=["2019"])["2019"].columns

dfResult = pd.DataFrame(columns=column_array)
```

```python
# Iterate through each sheet in the Excel file
for sheet_name in xls.sheet_names:
     
    # Check if the sheet name is not in the specified list
    if sheet_name not in ['1996', '1997', '1998', '1999', '2000', '2001']:
        # Read the sheet into a DataFrame
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        # print(sheet_name)
        # print(df.shape)
        # print(df.columns)
        
        # Ensure all required columns are present, add missing ones as None
        for column in column_array:
            if column not in df.columns:
                df[column] = None
                
        # Rename and split specific columns based on their names
        for column in df.columns:
            if column == "BUTTERMILK CENTRAL":
                df["SWEET CREAM BUTTERMILK CENTRAL"] = df[column]
                df.drop(columns=[column], inplace=True)
            elif column == "NONFAT CENTRAL/EAST":   
                df["NONFAT CENTRAL"] = df[column]
                df["NONFAT EAST"] = df[column]
                df.drop(columns=[column], inplace=True)
            elif column == "NONFAT CENTRAL  EAST":
                df["NONFAT CENTRAL EAST"] = df[column]
                df.drop(columns=[column], inplace=True)
            elif column == "CENTRAL EAST":
                df["NONFAT CENTRAL EAST"] = df[column]
                df.drop(columns=[column], inplace=True)
            elif column == "NDPSRNFDM":
                df["NDPSR NFDM"] = df[column]
                df.drop(columns=[column], inplace=True)
                
        # Add a 'Year' column based on the sheet name
        df['Year'] = sheet_name
        print("Rows to add:", df.shape[0], "from sheet", sheet_name, "of the Excel file")
        
        # Concatenate the current DataFrame to the result DataFrame
        dfResult = pd.concat([dfResult, df], ignore_index=True, sort=False)
 
# Display the first few rows of the result DataFrame
dfResult.head()

# Print the shape and sorted column names of the result DataFrame
print(dfResult.shape)
print(dfResult.columns.sort_values())
```
